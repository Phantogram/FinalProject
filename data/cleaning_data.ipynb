```python
import pandas as pd
```

```python
dinner_df = pd.read_csv("dinner_track.csv")
```

```python
party_df = pd.read_csv("party_track.csv")
```

```python
sleep_df = pd.read_csv("sleep_track.csv")
```

```python
workout_df = pd.read_csv("workout_track.csv")
```

```python
dinner_df = dinner_df.assign(Group = 'dinner')
party_df = party_df.assign(Group='party')
sleep_df = sleep_df.assign(Group='sleep')
workout_df = workout_df.assign(Group='workout')
```

```python
dinner_df = dinner_df.assign(group_n = 0)
party_df = party_df.assign(group_n=1)
sleep_df = sleep_df.assign(group_n=2)
workout_df = workout_df.assign(group_n=3)
concated = pd.concat([dinner_df, party_df,sleep_df,workout_df])
```

```python
concated = pd.concat([dinner_df, party_df,sleep_df,workout_df])
```

```python
concated = concated.dropna()
```

```python
concated = concated.drop_duplicates(subset=['id'])
```
